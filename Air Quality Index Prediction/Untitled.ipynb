```python
import pandas as pd

df = pd.read_csv('Processing_01.csv')
```

```python
df
```

```python
# One hot encoding for Site Name
df_SN_OHE = pd.get_dummies(df['Site Name (of Overall AQI)'], dtype = int)

# Drop one column to avoid dummies variable trap
df_SN_OHE = df_SN_OHE.drop(['YORK (LANDFILL)'], axis = 'columns')
df_SN_OHE
```

```python
# One Hot Encoding for city
df_city_OHE = pd.get_dummies(df['city'], dtype = int)

# Drop one column to avoid dummies variable trap
df_city_OHE = df_city_OHE.drop(['Washington'], axis = 'columns')
df_city_OHE
```

```python
# One Hot Encoding for state
df_state_OHE = pd.get_dummies(df['state'], dtype = int)

# Drop one column to avoid dummies variable trap
df_state_OHE = df_state_OHE.drop(['WA'], axis = 'columns')
df_state_OHE
```

```python
df_OHE = pd.concat([df, df_SN_OHE, df_city_OHE, df_state_OHE], axis = 'columns')
df_OHE
```

```python
# Drop the Site Name, city and state collumns
df_OHE.drop(['Site Name (of Overall AQI)', 'city', 'state'], axis = 1, inplace = True)
```

```python
df_OHE
```

```python
df_OHE.to_csv('OHE_data.csv', index = False)
```
