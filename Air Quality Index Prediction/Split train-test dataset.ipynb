```python
import pandas as pd

df = pd.read_csv('./data/OHE.csv')
```

```python
from sklearn.model_selection import train_test_split
```

```python
df
```

```python
hazardous_idx = df[df['AQI category'] == 'Hazardous'].index
df.drop(hazardous_idx, inplace = True)
```

```python
df['AQI category'].value_counts()
```

```python
#seperate features and target
Y = df['AQI category']
X = df.drop(['AQI category'], axis = 'columns')
```

```python
#split train-test dataset
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# y_val, x_val to csv files
x_train.to_csv('./data/x_train_dataset.csv', index = False)
y_train.to_csv('./data/y_train_dataset.csv', index = False)
x_test.to_csv('./data/x_test_dataset.csv', index = False)
y_test.to_csv('./data/y_test_dataset.csv', index = False)
```
