```python
import pandas as pd

df = pd.read_csv('./data/aqidataset.csv')
```

```python
df
```

```python
# Replace '.' with None
df.replace(to_replace = '.', value = None, inplace = True)
```

```python
# Remove Empty Cells
df.dropna(inplace = True)
```

```python
df
```

```python
# Remove ID-type columns
df.drop(['Unnamed: 0', 'Site ID (of Overall AQI)', 'Source (of Overall AQI)', 'Main Pollutant'], axis=1, inplace = True)
```

```python
df
```

```python
#Reset index of df
df.reset_index(drop=True, inplace=True)
```

```python
df['Date']
```

```python
# Seperate Date Month Year
date = []
month = []
year = []

for Date in df['Date']:
    date.append(Date[0:2])

for Month in df['Date']:
    month.append(Month[3:5])

for Year in df['Date']:
    year.append(Year[6:10])
```

```python
dict = {'Date': date, 'Month': month, 'Year': year}

dict_df = pd.DataFrame(dict)
dict_df
```

```python
df.drop(['Date'], axis = 1, inplace = True)
df = pd.concat([df, dict_df], axis = 1)
```

```python
# Remove Ducplicate
df.drop_duplicates(inplace = True)
```

```python
df
```

```python
#Change type of some columns to float64
df[['Overall AQI Value', 'Ozone', 'PM25', 'CO', 'PM10', 'NO2', 'Date', 'Month', 'Year']] = df[['Overall AQI Value', 'Ozone', 'PM25', 'CO', 'PM10', 'NO2', 'Date', 'Month', 'Year']].astype('float')
```

```python
#Change type of some columns to int64
df[['Overall AQI Value', 'Ozone', 'PM25', 'CO', 'PM10', 'NO2', 'Date', 'Month', 'Year']] = df[['Overall AQI Value', 'Ozone', 'PM25', 'CO', 'PM10', 'NO2', 'Date', 'Month', 'Year']].astype('int64')
```

```python
df.info()
```

```python
df.to_csv('./data/Processing_01.csv', index = False)
```
