```python
import pandas as pd

df = pd.read_csv('aqidataset.csv')
```

```python
df
```

```python
# Replace '.' with None
df.replace(to_replace = '.', value = None, inplace = True)
```

```python
# Remove Empty Cells
df.dropna(inplace = True)
```

```python
df
```

```python
# Remove ID-type columns
df.drop(['Unnamed: 0', 'Site ID (of Overall AQI)', 'Source (of Overall AQI)', 'Main Pollutant'], axis=1, inplace = True)
```

```python
df
```

```python
df['Date']
```

```python
# Keep only the year in Date
for date in df['Date']:
    df['Date'] = date[6:10]
```

```python
df
```

```python
# Remove Ducplicate
df.drop_duplicates(inplace = True)
```

```python
df
```

```python
#Change type of some columns to float64
df[['Date', 'Overall AQI Value', 'Ozone', 'PM25', 'CO', 'PM10', 'NO2']] = df[['Date', 'Overall AQI Value', 'Ozone', 'PM25', 'CO', 'PM10', 'NO2']].astype('float')
```

```python
#Change type of some columns to int64
df[['Date', 'Overall AQI Value', 'Ozone', 'PM25', 'CO', 'PM10', 'NO2']] = df[['Date', 'Overall AQI Value', 'Ozone', 'PM25', 'CO', 'PM10', 'NO2']].astype('int64')
```

```python
df.info()
```

```python
df.to_csv('Processing_01.csv', index = False)
```
